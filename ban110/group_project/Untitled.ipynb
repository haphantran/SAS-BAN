# BAN110 Group project - Heart disease dataset



```sas
*Reading dataset;
libname mylib '/folders/myfolders/ban110/group_project';

FILENAME REFFILE '/folders/myfolders/ban110/group_project/heart_Kaggle.csv';

PROC IMPORT DATAFILE=REFFILE
    DBMS=CSV
    OUT= mylib.heart;
    GETNAMES=YES;
RUN;
*label the dataset;
data mylib.heart;
    set mylib.heart;
    label
        age ="age in years"
        sex="Gender (1 Male, 0 Female)"
        cp="chest pain type (4 values)"
        trestbps = "resting blood pressure (in mm Hg on admission to the hospital)"
        chol="serum cholestoral in mg/dl"
        fbs="(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)"
        restecg="restecg: resting electrocardiographic results 
                (0: normal, 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
                ,2: showing probable or definite left ventricular hypertrophy by Estes' criteria)"
        thalach="maximum heart rate achieved"
        exang="exercise induced angina (1 = yes; 0 = no)"
        oldpeak="ST depression induced by exercise relative to rest"
        slope="the slope of the peak exercise ST segment (1 Upsloping, 2 flat, 3 downsloping)"
        ca="number of major vessels (0-3) colored by flourosopy"
        thal="3 = normal; 6 = fixed defect; 7 = reversable defect"
        target= "diagnosis of heart disease (angiographic disease status)
-- Value 0: < 50% diameter narrowing
-- Value 1: > 50% diameter narrowing"
run;

*create id field;
data mylib.heart;
    retain id;
    set mylib.heart;
    id = _n_;
run;

title "First 15 rows from dataset";
proc print data=mylib.heart (obs=15);
run;
```

## Dataset characteristic

```sas
title "Contents of the dataset";
proc contents data=mylib.heart order=varnum;
run;

title "Frequency table of target variable";
proc freq data=mylib.heart;
    tables target /  nocum;
run;
```

2. List of variables names: output of proc contents. Type: All numeric
3. Target varaible: target. Dependent variable: all other variables
4. The dataset is balanced between 2 values 0 and 1 of target variable

### Check for missing values

```sas
proc format;
    value MissingCheck
            . = 'Missing'
        other = 'non-missing';
run;

proc freq data=mylib.heart;
    tables thal /  nocum;
    format thal MissingCheck.;
run;
*run same step for differnt variable to find missing value
```

```sas
*wait for missing data;
proc stdize data=Have out=Imputed 
      oprefix=Orig_         /* prefix for original variables */
      reponly               /* only replace; do not standardize */
      method=mode;          /* or MEDIAN, MINIMUM, MIDRANGE, etc. */
   var Height;              /* you can list multiple variables to impute */
run;
```

### Derived variable
ca="number of major vessels (0-3) colored by flourosopy"
ca_new = 0, if ca = 0
        1, if ca>0

```sas
data  mylib.heart;
    set mylib.heart;
    if ca = 0 then ca_new = 0;
    else ca_new = 1;
run;

proc freq data=mylib.heart;
    tables ca ca_new /nocum;
run;
```

#### Q3. Examine the variable "education" 
    3.1. define a new format, name it education_Check and use it to identify invalid values for the variable education. 
    Valid values are 'primary', 'secondary', 'tertiary', 'unknown'. 

    Refer to program 1.8. 
    Chapter 1 - Working with Character Data
    Cody's Data Cleaning Techniques Using SAS, Third Edition

    

```sas
proc format; 
 value $education_Check 'primary', 'secondary', 'tertiary', 'unknown'= 'valid'
                       ' '   = ' missing'
                       other = 'Error'; 
run;

proc freq data=mylib.customer_all; 
    tables education / nocum nopercent missing; 
    format education $education_Check.; 
run; 
```

    3.2. Use the function lowcase on education column. use the same dataset name for output dataset.

```sas
data mylib.customer_all;
    set mylib.customer_all;
    education = lowcase(education);
run;
```

    3.3. show the simple frequency table after the change. 

```sas
proc freq data=mylib.customer_all; 
    tables education / nocum nopercent missing; 
    format education $education_Check.; 
run; 
```

#### Q4.  Examine the variable "marital". 
    4.1. Use PROC print with a where statement to check for data errors in the variable marital. Consider the valid values as "single", "married", "divorced". 
    
    Refer to program 1.6.  
    Chapter 1 - Working with Character Data
    Cody's Data Cleaning Techniques Using SAS, Third Edition


```sas

title "Using PROC Print to Identify Data Errors in variable marital";

proc print data=mylib.customer_all; 
    id customer_id; 
    var marital; 
    where marital not in ('single', 'married', 'divorced');
run;
```

    4.2. Use the function lowcase on the variable marital.

```sas
data mylib.customer_all; 
    set mylib.customer_all; 
    marital = lowcase(marital);
run;
```

    4.3. show the simple frequency table after the change. 

```sas
title "Frequency table for variable marital after lower case";
proc freq data=mylib.customer_all; 
    tables marital / nocum nopercent missing; 
run;
```

#### Q7. create a new variable named jobMF to indicate the most frequent job category
Reuse the code provided in ch17, section 17.3.2. 
- check the most frequent job category based on the output of proc freq. 
- create the new variable jobMF
- print the first few observations. 

```sas
proc freq data=mylib.customer_all order=freq; 
    table job;
run; 

data mylib.customer_all; 
    set mylib.customer_all; 
    IF job = 'management' THEN jobmf = 1;
    ELSE jobmf =0;
run; 

proc print data=mylib.customer_all (obs=5); 
run;
```

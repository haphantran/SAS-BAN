#### 2.2. write a piece of code to output the customer_id whose age is missing. Use the function missing within if statements


```sas
data customer_id_missing_age (keep = customer_id);
    set mylib.customer_all;
    if missing(age);
run;
title "Customer_id with missing age";
proc print data = customer_id_missing_age;
    id customer_id;
run;
```

#### 2.3. apply imputation to replace missing values for age with the mean age. 
Name the imputed dataset as customer_all_Imputed. 
     
For reference, use PROC STDIZE to replace missing values with mean.   
Here is a sample code:

    /* Mean imputation: Use PROC STDIZE to replace missing values with mean */
    
    proc stdize data=Have out=Imputed 
          oprefix=Orig_         /* prefix for original variables */
          reponly               /* only replace; do not standardize */
          method=MEAN;          /* or MEDIAN, MINIMUM, MIDRANGE, etc. */
       var Height;              /* you can list multiple variables to impute */
    run;

    proc print data=Imputed;
       format Orig_Height Height BESTD8.1;
       var Name Orig_Height Height Weight Replaced;
    run;


```sas
proc stdize data=mylib.customer_all out=mylib.customer_all_Imputed 
      oprefix=Orig_         /* prefix for original variables */
      reponly               /* only replace; do not standardize */
      method=MEAN;          /* or MEDIAN, MINIMUM, MIDRANGE, etc. */
   var age;              /* you can list multiple variables to impute */
run;

title "customer_id with imputed age using mean";
proc print data=mylib.customer_all_Imputed;
   var customer_id Orig_age age;
   where missing(orig_age);
run;
```

#### 3.1.  Binning  
- discretize the variable age by creating a new cat variable named age_cat: 
        18 <=AGE <=35 for young_adult,  
        36 <=AGE <=55 for middle_age  
        >56 for old.  
- show a simple frequency table for age_cat


```sas
data mylib.customer_all_imputed_binned; 
     set mylib.customer_all_imputed; 
     format age_cat $15.; 
     
     if age ge 18 and age le 35 then age_cat = 'young_adult'; 
     else if age le 55 then age_cat = 'middle_age'; 
     else  age_cat= 'old' ;     
run; 
title "Age_cat frequency table";
proc freq data=mylib.customer_all_imputed_binned;
    tables age_cat;
run; 
```

####  Here is the code to create a bar chart of balance by age.  


```sas
proc sgplot data=mylib.customer_all_imputed_binned;
    vbar age_cat / response=balance group=balance groupdisplay=cluster stat=mean;
    yaxis grid;
run;
```

### 4. Examine the variable campaign 
campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

#### 4.1. Use proc univariate on campaign. 



```sas
title "Running PROC UNIVARIATE on Campaign";
proc univariate data=mylib.customer_all nextrobs=5;
   id customer_id;
   var campaign;
   histogram / normal;
run;
```

#### 4.2. Interpret the quantiles table for the variable Campaign. 

- The min value for campaign is 1.
- The first quartile is 1 means that 25% of the dataset have the value of campaign less than or equal to 1.
- The second quartile (median) is 2 means that 50% of the dataset have the value of campaign less than or equal to 2.
- The third quartile is 3 means that 75% of the dataset have the value of campaign less than or equal to 3.
- The max value for campaign is 50.


#### 4.3. Based on quantiles table, the variable campaign seems  more categorical in nature than continuous. 
Convert the variable campaign into a categorical variable name "campaign_cat" with ordinal values { 1, 2, 3, >3} 


```sas
data mylib.customer_all_new; 
     set mylib.customer_all_imputed_binned; 
     format campaign_cat $3.; 
     
     if campaign > 3 then campaign_cat = '>3'; 
     else campaign_cat = put(campaign, 3.);
run; 
title "Campaign_cat frequency table";
proc freq data=mylib.customer_all_new;
    tables campaign_cat ;
run; 

```

### 5. Examine  the variable pdays 
pdays: number of days that passed by after the client was last contacted from a previous campaign 
 (numeric; -1 means client was not previously contacted)
 
#### 5.1. Use proc univariate on the variable pdays. 


```sas
title "Running PROC UNIVARIATE on pdays";
proc univariate data=mylib.customer_all_new nextrobs=5;
   id customer_id;
   var pdays;
   histogram / normal;
run;
```

#### 5.2. creating a derived variable
By checking the quantiles table in the proc univariate output, clearly it is better to create a new categorical variable named "contacted_before" that takes the value  'yes' if the customer has been contacted before and 'no' if the customer was not contacted before in a previous campaign (pdays=-1) 


```sas
data mylib.customer_all_new1;
    set mylib.customer_all_new;
    format contacted_before $3.; 
    if pdays = -1 then contacted_before = "no";
        else contacted_before = "yes";
run;

title "Frequency table for new contacted_before variable";
proc freq data = mylib.customer_all_new1;
    tables contacted_before;
run;
```

#### 5.3. print the first 5 observations where pdays>0

```sas
title "First 5 observations where pdays>0";
proc print data = mylib.customer_all_new1 (obs = 5);
    where pdays>0;
run;
```

#### 5.4. drop the column pdays 

```sas
data mylib.customer_all_new;
    set mylib.customer_all_new1 (drop = pdays);    
run;
```
